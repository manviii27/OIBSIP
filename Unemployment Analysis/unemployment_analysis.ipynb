# Graph showing unemployment rate of each state

```python
colors = plt.cm.tab20c(np.arange(27))
```

```python
plt.bar(data['Region'].unique(),df[' Estimated Unemployment Rate (%)'],color=colors,label=data['Region'].unique())
plt.xticks(data['Region'].unique(), rotation='vertical')
plt.ylabel('Estimated Unemployment Rate')
plt.title('Average Unemployment Rate in each state')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left', fontsize=8)
plt.show()
```
